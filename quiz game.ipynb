Using Python, develop a quiz game that asks users multiple-choice or fill-in-the-blank questions on a specific topic. The game should keep track of scores, provide feedback on correct/incorrect answers, and offer a variety of questions to make it challenging and engaging. 
Load Quiz Questions.
Display Welcome Message and Rules.
Present Quiz Questions:

Display each question and answer choice.
Prompt the user to select an answer.
Evaluate the User's Answer:

Compare the user's answer with the correct answer.
Keep track of the user's score.
Provide Feedback:

Display if the answer was correct or incorrect.
Show the correct answer for incorrect responses.
Calculate the final score.
Display Final Results:

Print the userâ€™s score and performance message. 
Play again: 
Ask the user if they want to play again

```python
import random

questions = [
    {
        "question": "What is the capital of Nigeria?",
        "options": ["London", "Berlin", "Abuja", "Madrid"],
        "correct_answer": "Abuja",
    },
    {
        "question": "What is the full meaning of MMA?",
        "options": ["MY Main Agenda", "Missed Marshall Activity", "Mixed Marshall Arts", "Missed Marshall Arts"],
        "correct_answer": "Mixed Marshall Arts",
    },
    {   "question": "What is the color of blood?",
        "options": ["red", "yellow", "brown", "purple"],
        "correct_answer": "red"}]

def ask_question(question):
    print(question["question"])
    for i, option in enumerate(question["options"], 1):
        print(f"{i}. {option}")
    user_answer = input("Enter the number of your answer: ")
    return user_answer

def check_answer(question, user_answer):
    return question["options"][int(user_answer) - 1] == question["correct_answer"]

def play_quiz():
    score = 0
    for question in random.sample(questions, len(questions)):
        user_answer = ask_question(question)
        if check_answer(question, user_answer):
            print("Correct!\n")
            score += 1
        else:
            print(f"Sorry, the correct answer is {question['correct_answer']}.\n")
    return score

def display_results(score, total_questions):
    print(f"Your Score: {score}/{total_questions}")
    if score == total_questions:
        print("Congratulations! You got a perfect score!")
    elif score >= total_questions // 2:
        print("Good job! You did well.")
    else:
        print("You can do better. Keep practicing!")

while True:
    print("Welcome to the Quiz Game!")
    play_again = input("Would you like to play? (yes/no): ")
    if play_again.lower() != "yes":
        break
total_questions = len(questions)
score = play_quiz()
display_results(score, total_questions)

play_again = input("Play again? (yes/no): ")
if play_again.lower() != "yes":
        break
```
